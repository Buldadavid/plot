```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('matlab.mplstyle')
```

```python
# 1. Definice parametrů signálu
f = 5          # Frekvence signálu v Hz (kolikrát se sinusovka zopakuje za sekundu)
A = 0.5          # Amplituda signálu (maximální výška od nuly)
phi = 0        # Fázový posun v radiánech (posun v čase doleva/doprava)
dc = 1         # Stejnosměrný posun / DC offset (posun signálu nahoru/dolů)

# 2. Definice parametrů vzorkování (časová osa)
fs = 1000      # Vzorkovací frekvence v Hz (počet bodů za sekundu)
t = np.linspace(0,10,10*fs)
```

```python
#vše OK
y = A * np.sin(2 * np.pi * f * t + phi) + 0
y2 = A * np.sin(2 * np.pi * f * t - np.pi/2) + 0

fig = plt.subplots(figsize=(5,2.5), layout='constrained')
plt.plot(t, y)
plt.plot(t, y2)
plt.xlim([0,0.5])

fig = plt.subplots(figsize=(5,5), layout='constrained')

plt.scatter(y,y2, linewidths= 0.1, marker=".", alpha=0.25)
plt.xlim(-0.67,0.67)
plt.ylim(-0.67,0.67)
plt.vlines(0,-0.67,0.67,color = 'k', linestyles='--',lw = 0.75)
plt.hlines(0,-0.67,0.67,color = 'k', linestyles='--',lw = 0.75)
plt.show()
```

```python
#chyba - měnící se amplituda = rozšiřuje kružnici (opotřebení ložisek - Mění se vzdálenost mezi skleněným diskem símačem / znečištění)
A = 0.5 + (0.025 * np.sin(2 * np.pi * f/10 * t + phi) + 0)

y = A * np.sin(2 * np.pi * f * t + phi) + 0
y2 = A * np.sin(2 * np.pi * f * t - np.pi/2) + 0


fig = plt.subplots(figsize=(5,2.5), layout='constrained')
plt.plot(t, y)
plt.plot(t, y2)
plt.xlim([0,5.5])

fig = plt.subplots(figsize=(5,5), layout='constrained')

plt.scatter(y,y2, linewidths= 0.1, marker=".", alpha=0.25)
plt.xlim(-0.67,0.67)
plt.ylim(-0.67,0.67)
plt.vlines(0,-0.67,0.67,color = 'k', linestyles='--',lw = 0.75)
plt.hlines(0,-0.67,0.67,color = 'k', linestyles='--',lw = 0.75)
plt.show()
```

```python
#chyba - měnící se frekvence = nic (prí konstantních 1000ot/min - vibrace, prokluz hřídelky, povolené praporky?)

k = (5 - 1) / 1 # Rychlost změny frekvence
f0 = 5

# Pozor: Při integraci frekvence pro fázi vzniká u lineární změny činitel t^2 / 2
y = 0.5 * np.sin(2 * np.pi * (f0 * t + 0.5 * k * t**2) + phi) + 0
y2 = 0.5 * np.sin(2 * np.pi * (f0 * t + 0.5 * k * t**2) - np.pi/2) + 0

fig = plt.subplots(figsize=(5,2.5), layout='constrained')
plt.plot(t, y)
plt.plot(t, y2)
plt.xlim([0,5.5])

fig = plt.subplots(figsize=(5,5), layout='constrained')

plt.scatter(y,y2, linewidths= 0.1, marker=".", alpha=0.25)
plt.xlim(-0.67,0.67)
plt.ylim(-0.67,0.67)
plt.vlines(0,-0.67,0.67,color = 'k', linestyles='--',lw = 0.75)
plt.hlines(0,-0.67,0.67,color = 'k', linestyles='--',lw = 0.75)
plt.show()
```

```python
#chyba - DC posun jedne se sin = vychýlení od středu (opotřebovaná LED nebo snímač, vibrace, špína)

y = 0.5 * np.sin(2 * np.pi * f * t + phi) + 0
y2 = 0.5 * np.sin(2 * np.pi * f * t - np.pi/2) + 0.15

fig = plt.subplots(figsize=(5,2.5), layout='constrained')
plt.plot(t, y)
plt.plot(t, y2)
plt.xlim([0,0.5])

fig = plt.subplots(figsize=(5,5), layout='constrained')

plt.scatter(y,y2, linewidths= 0.1, marker=".", alpha=0.25)
plt.xlim(-0.67,0.67)
plt.ylim(-0.67,0.67)
plt.vlines(0,-0.67,0.67,color = 'k', linestyles='--',lw = 0.75)
plt.hlines(0,-0.67,0.67,color = 'k', linestyles='--',lw = 0.75)
plt.show()
```

```python
#chyba - fázový posun != 90° = natočení kruhu do elipsy (špatní geometrie upenění snímače)

phi = np.linspace(np.pi/2,np.pi/2.2,10000)

y = 0.5 * np.sin(2 * np.pi * f * t + 0) + 0
y2 = 0.5 * np.sin(2 * np.pi * f * t - phi) + 0
y3 = 0.5 * np.sin(2 * np.pi * f * t + 0) + 0
y4 = 0.5 * np.sin(2 * np.pi * f * t - np.pi/2.2) + 0

fig = plt.subplots(figsize=(5,2.5), layout='constrained')
plt.plot(t, y)
plt.plot(t, y2)
plt.xlim([0,0.5])

fig = plt.subplots(figsize=(5,5), layout='constrained')

plt.scatter(y,y2, linewidths= 0.1, marker=".", alpha=0.25, label = "fáz. posun 82-90°")
plt.scatter(y3,y4, linewidths= 0.1, marker=".", alpha=0.25, label ="fáz. posun je 82°")
plt.xlim(-0.67,0.67)
plt.ylim(-0.67,0.67)
plt.vlines(0,-0.67,0.67,color = 'k', linestyles='--',lw = 0.75)
plt.hlines(0,-0.67,0.67,color = 'k', linestyles='--',lw = 0.75)
plt.legend()
plt.show()
```
