```python
import numpy as np
from scipy.fft import fft, fftfreq
import matplotlib.pyplot as plt


def propoj_vrcholy_sinusovky(frekvence, vzorkovaci_frekvence, sinusoida, casova_osa):
    min_vzdalenost_mezi_vrcholy = int(vzorkovaci_frekvence / frekvence * 0.8) # 0.8 pro jistotu
    vrcholy_indexy, _ = find_peaks(sinusoida, distance=min_vzdalenost_mezi_vrcholy)
    # Získání hodnot časové osy a amplitudy pro nalezené vrcholy
    cas_vrcholu = casova_osa[vrcholy_indexy]
    amplituda_vrcholu = sinusoida[vrcholy_indexy]
    return cas_vrcholu, amplituda_vrcholu

def measure_frequency_of_sine_wave(signal_data, sample_rate):

    N = len(signal_data)  # Počet vzorků
    yf = fft(signal_data) # Provede rychlou Fourierovu transformaci
    xf = fftfreq(N, 1 / sample_rate) # Vypočítá frekvenční osy

    # Vezmeme pouze kladné frekvence (spektrum je symetrické)
    # Ignorujeme frekvenci 0 Hz (stejnosměrnou složku)
    positive_freq_indices = np.where(xf > 0)
    xf_positive = xf[positive_freq_indices]
    yf_positive = np.abs(yf[positive_freq_indices])

    # Najdeme index frekvence s největší amplitudou
    dominant_frequency_index = np.argmax(yf_positive)
    dominant_frequency = xf_positive[dominant_frequency_index]

    return dominant_frequency

# Výpočet frekvence

sample_rate = 1/350e-9  # Vzorkovací frekvence v Hz (např. 1000 vzorků za sekundu)
duration = tt[-1]      # Doba trvání signálu v sekundách

# --- Měření frekvence ---
measured_freq = measure_frequency_of_sine_wave(a, sample_rate)
print(f"Změřená frekvence signálu: {measured_freq:.2f} Hz")
cas_vrcholu, amplituda_vrcholu = propoj_vrcholy_sinusovky(measured_freq, sample_rate, a, t)

# --- Volitelná vizualizace ---

plt.figure(figsize=(12, 9))

plt.subplot(3, 1, 1)
plt.plot(tt,a, label="sin A")
plt.title('Sinusoida A v časové doméně')
plt.xlabel('Čas [ms]')
plt.ylabel('Amplituda')
plt.grid(True)
plt.xscale("log")
plt.xlim(0.01,2)
plt.hlines(0.375,0,cas_vrcholu[-1],color = 'g',lw = 0.75)
plt.hlines(0.6,0,cas_vrcholu[-1],color = 'g',lw = 0.75)
plt.hlines(-0.375,0,cas_vrcholu[-1],color = 'g',lw = 0.75)
plt.hlines(-0.6,0,cas_vrcholu[-1],color = 'g',lw = 0.75)
#plt.legend(bbox_to_anchor=(0,1,1,.1), ncol=2, mode = "expand", loc = "lower left")

# Zobrazení frekvenčního spektra

Na = len(a)
Ta = 1/sample_rate
yfa = fft(a)
xfa = fftfreq(Na, Ta)[:Na//2]


plt.subplot(3, 1, 2)
plt.plot(xfa, 2.0/Na * np.abs(yfa[0:Na//2]))
plt.title('Frekvenční spektrum (FFT)')
plt.xlabel('Frekvence [Hz]')
plt.ylabel('Amplituda (normalizovaná)')
plt.grid(True)
# plt.xlim(0, measured_freq * 2) # Omezíme zobrazení na zajímavou část spektra
plt.xlim(measured_freq - 1000, measured_freq + 1000) # Omezíme zobrazení na zajímavou část spektra
tloustka = max(otacky) - min(otacky)
plt.vlines(34254.737,0,max( 2.0/Na * np.abs(yfa[0:Na//2])),color = 'g',lw = tloustka, alpha=0.25)
#plt.vlines(34285.044,0,max( 2.0/Na * np.abs(yfa[0:Na//2])),color = 'g',lw = 2)

plt.subplot(3, 1, 3)
#plt.plot(tt, a, label='Původní sinusovka')
# Spojení vrcholů čarami
plt.plot(cas_vrcholu, amplituda_vrcholu, label='Spojené vrcholy',lw = 0.75)
plt.hlines(0.375,0,cas_vrcholu[-1],color = 'g',lw = 0.75)
plt.hlines(0.6,0,cas_vrcholu[-1],color = 'g',lw = 0.75)
plt.title(f'Sinusovka A s propojenými vrcholy (Frekvence: {measured_freq:.2f} Hz)')
plt.xlabel('Čas [ms]')
plt.ylabel('Amplituda')
#plt.legend()
plt.ylim(0.3,0.7)
plt.xlim(0,tt[-1])
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def check_phase_shift(signal1, signal2, fs, target_phase_deg=90, tolerance_deg=5):
    """
    Zkontroluje, zda fázový posun mezi dvěma signály je v rozsahu
    cílové fáze +/- tolerance.

    Parametry:
    - signal1 (np.array): První sinusový průběh (referenční).
    - signal2 (np.array): Druhý sinusový průběh.
    - fs (int): Vzorkovací frekvence signálů.
    - target_phase_deg (float): Požadovaný fázový posun ve stupních (výchozí 90).
    - tolerance_deg (float): Povolená tolerance ve stupních (výchozí 5).

    Vrací:
    - bool: True, pokud fázový posun splňuje kritéria, jinak False.
    - float: Odhadovaný fázový posun ve stupních.
    """

    # 1. Křížová korelace k nalezení posunu
    # Normalizovaná křížová korelace
    corr = np.correlate(signal1, signal2, mode='full')
    max_corr_index = np.argmax(corr)

    # Index posunu je vzdálenost od středu korelačního pole
    delay_samples = max_corr_index - (len(signal1) - 1)

    # 2. Převod posunu na čas a fázi
    delay_time = delay_samples / fs

    # Odhad frekvence signálu (pro zjednodušení předpokládejme dominantní frekvenci)
    # Pro přesný výpočet fázového posunu potřebujeme znát frekvenci signálu
    # Použijeme FFT k odhadu hlavní frekvence.
    N = len(signal1)
    T = 1.0 / fs
    yf = np.fft.fft(signal1)
    xf = np.fft.fftfreq(N, T)[:N//2]
    # Najdeme index dominantní frekvence
    dominant_freq_index = np.argmax(np.abs(yf[1:N//2])) + 1
    f_signal = xf[dominant_freq_index]

    if f_signal == 0:
        return False, 0.0

    # Fázový posun v radiánech: posun v čase * 2*pi * frekvence
    phase_shift_rad = 2 * np.pi * f_signal * delay_time
    phase_shift_deg = np.degrees(phase_shift_rad)

    # Normalizace fázového posunu do rozsahu -180 až 180 stupňů
    normalized_phase_shift = (phase_shift_deg + 180) % 360 - 180

    # 3. Kontrola tolerance
    lower_bound_p = target_phase_deg - tolerance_deg  # 85
    upper_bound_p = target_phase_deg + tolerance_deg  # 95
    # Také kontrola ekvivalentního negativního posunu (např. -270 místo +90)
    lower_bound_n = target_phase_deg - 360 - tolerance_deg  # -275
    upper_bound_n = target_phase_deg - 360 + tolerance_deg  # -265

    is_in_range = (lower_bound_p <= normalized_phase_shift <= upper_bound_p) or \
                  (lower_bound_n <= normalized_phase_shift <= upper_bound_n)

    return is_in_range, normalized_phase_shift


result_out, phase_out = check_phase_shift(b, a, 1/350e-9)
print(f"Odhadovaný fázový posun: {phase_out:.2f} stupňů")
print(f"Je v rozsahu 90° ± 5°: {result_out}")

# --- Volitelné: Vykreslení pro vizualizaci ---
plt.figure(figsize=(10, 4))
plt.plot(tt[:100], a[:100], label='Signál 1 (Ref)')
plt.plot(tt[:100], b[:100], label=f'Signál 2 ')
plt.title('Průběhy signálů')
plt.xlabel('Čas [s]')
plt.ylabel('Amplituda')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pywt

# Parametry
Fs = 5714.29  # Vzorkovací frekvence
T = 1 / Fs
L = 2858

# Časový vektor
t = np.arange(0, L) * T

# Sinusový signál
x = np.sin(2 * np.pi * 16.7*2 * t)

# Vykreslení časového průběhu
plt.figure(figsize=(10, 4))
plt.plot(t, x)
plt.title("Sinusový signál (16.7 Hz)")
plt.xlabel("Čas (s)")
plt.ylabel("Amplituda")
plt.grid(True)
plt.show()

# CWT pomocí PyWavelets
scales = pywt.scale2frequency('morl', np.arange(1, 100 + 1)) / (1.0/Fs)
coeffs, freqs = pywt.cwt(x, scales, 'morl', sampling_period=T)

# Vykreslení spektrogramu
plt.figure(figsize=(10, 4))
#plt.imshow(np.abs(coeffs)**2, extent=[t[0], t[-1], freqs[-1], freqs[0]],
#           aspect='auto', cmap='jet')

plt.imshow(np.abs(coeffs)**2, aspect='auto', 
           extent=[t.min(), t.max(), freqs.min(), freqs.max()],
           cmap='jet', origin='lower') # <--- Tato část je klíčová

plt.colorbar(label='Energie')
plt.title("Spektrogram (CWT)")
plt.xlabel("Čas (s)")
plt.ylabel("Frekvence (Hz)")
plt.ylim([0.1, 50])
plt.show()
```
