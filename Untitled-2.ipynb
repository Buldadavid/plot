```python
#osciloskop

import matplotlib.pyplot as plt
import numpy as np
import scipy

data = np.genfromtxt(r'C:\Users\z003zh2j\Desktop\PLOT\nic2_7.csv', delimiter=',', skip_header=1)

t = data[: , 0]
R = (data[: , 1]/50)+5.7
U = data[: , 2]*2
V = data[: , 3]
W = data[: , 4]


fig = plt.subplots(figsize=(13,7), layout='constrained')
plt.plot(t,U,'r',lw = 1, label = "U")
plt.plot(t,V,'b',lw = 1, label = "V")
plt.plot(t,W,'k',lw = 1, label = "W")
plt.plot(t,R,'lime',lw = 1, label = "R")
plt.grid(color='lightgrey', linestyle='-', linewidth=0.5)
plt.ylim(-8, 8)
plt.xlim(-20 ,20)
plt.legend()
plt.xlabel('Čas (ms)')
plt.ylabel('napětí (mV)')
plt.show()
```
