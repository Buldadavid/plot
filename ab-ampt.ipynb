```python
import matplotlib.pyplot as plt
import numpy as np
import csv
import scipy
from scipy.signal import find_peaks

plt.style.use('moje.mplstyle')
```

```python
ABR = np.genfromtxt(r'S:\1.FILE\eqn\tycab.csv', delimiter=',', skip_header=1)

tt = ABR[: , 0]/1e6
a = ABR[: , 1]/1e3
b = ABR[: , 2]/1e3
```

```python
fig, host = plt.subplots(figsize=(13,7), layout='constrained')
ax = plt.gca()
plt.title("A,B na hrubo")
p1 = plt.plot(tt,a, label="sin a")
p2 = plt.plot(tt,b, label="sin b")
# plt.xlim(0,0.4)
host.set_xlabel("as [ms]")
host.set_ylabel("napt [V]")
host.legend(handles=p1+p2, loc='upper right')
plt.grid(color='lightgrey', linestyle='-', linewidth=0.5)

plt.show()
```

```python
#Oblka
from scipy.signal import hilbert
analytic_signal = hilbert(a)
amplitude_envelope = np.abs(analytic_signal)
X = amplitude_envelope
N = len(X)
n = np.arange(N)
T = 350e-9
freq = n/T 
plt.plot(freq, np.abs(X), lw=1)
plt.xlim(0,1e12)
plt.show()
```
