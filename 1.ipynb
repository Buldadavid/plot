```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
import scipy
```

```python
x = np.linspace(0,1,100)
y = np.sin(2*np.pi*x)

fig = plt.subplots(figsize=(13,7), layout='constrained')
ax = plt.gca()
ax.xaxis.set_major_locator(ticker.MultipleLocator(0.25))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(0.1))
ax.xaxis.set_ticks_position('bottom')
ax.tick_params(which='major', width=10.00, length=105)
ax.tick_params(which='minor', width=10.75, length=105, labelsize=100)
ax.tick_params(axis='x', color='m', length=10, direction='out', width=4, labelcolor='g', grid_color='b')
plt.plot(x,y)
plt.xlim(0,1)
plt.grid()
plt.show()
```
